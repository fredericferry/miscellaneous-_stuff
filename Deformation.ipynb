```python
%matplotlib inline

import os
import matplotlib.pyplot as plt
import numpy as np
from numpy import sqrt, pi, cos, sin, exp, meshgrid, linspace, arange

import warnings
warnings.filterwarnings('ignore')
```

```python
# Domaine -10 Ã  10
xx = linspace(-10,10,21)
yy = linspace(-10,10,21)
x,y = meshgrid(xx,yy)

# Domaine 0 Ã  10
xx1 = linspace(0,10,11)  
yy1 = linspace(0,10,11)
x1,y1 = meshgrid(xx1,yy1)
```

```python
alpha=3

u=alpha*x ; v=-alpha*y
#u=-alpha*x ; v=alpha*y
#u=alpha*y ; v=alpha*x
#u=-alpha*y ; v=alpha*x
#u=alpha*y1 ; v=0*x1
#u=alpha*x ; v=alpha*y
#u=-alpha*x ; v=-alpha*y
#u=alpha*x1 ; v=0*y1
#u=alpha*(1-x1) ; v=0*y1
```

```python
fig = plt.figure(figsize=(10, 10))

ax = fig.add_subplot(111)
ax.set_title('Horizontal wind field', fontsize=14)
plt.grid(True)
plt.xlabel("x", fontsize=10)
plt.ylabel("y", fontsize=10)
ax.quiver(x,y,u,v,sqrt(u**2+v**2), cmap="viridis")

figname='./wind'
plt.savefig(figname+'.png',bbox_inches='tight')

plt.show()
```
