```python
%matplotlib inline

import os
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
std_atm=np.loadtxt('./data/std_atm.txt',skiprows=1, usecols = (0,1,2,3))
alt=std_atm[:,0].flatten() # km
temp=std_atm[:,1].flatten() # °C
pres=10*std_atm[:,2].flatten() # hPa
density=std_atm[:,3].flatten() # kg/m3

print(std_atm)
print(alt)
```

```python
fig = plt.figure(figsize=(20,8))

fig.suptitle('US Standard atmosphere', )

ax = fig.add_subplot(131)
plt.title('Pressure vertical profile', fontsize=16)
plt.xlabel('Pressure (hPa)')
plt.ylabel('Altitude (km)')
plt.plot(pres, alt, color='red', linewidth=2)
plt.scatter(pres, alt, color='k', s=10)

ax = fig.add_subplot(132)
plt.title('Density vertical profile', fontsize=16)
plt.xlabel('Density (km/$m^3$)')
plt.ylabel('Altitude (km)')
plt.plot(density, alt, color='green', linewidth=2)
plt.scatter(density, alt, color='k', s=10)

ax = fig.add_subplot(133)
plt.title('Temperature vertical profile', fontsize=16)
plt.xlabel('Temperature (°C)')
plt.ylabel('Altitude (km)')
plt.scatter(temp, alt, color='k', s=10)
plt.plot(temp, alt, color='blue', linewidth=2)
plt.axhline(11, color='grey', linestyle="-", linewidth=0.5)
plt.axhline(50, color='grey', linestyle="-", linewidth=0.5)
plt.axhline(85, color='grey', linestyle="-", linewidth=0.5)

figname='./figs/std_atm'
plt.savefig(figname+'.png',bbox_inches='tight')
plt.show()
```
